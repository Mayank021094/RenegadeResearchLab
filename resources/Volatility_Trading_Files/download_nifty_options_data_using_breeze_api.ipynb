```python
def format_expiry_date(expiry_datetime):
    # 1) If they passed a string, parse it
    if isinstance(expiry_datetime, str):
        # adjust the format string to match your input format
        expiry_datetime = datetime.strptime(expiry_datetime, "%Y-%m-%d")
        expiry_datetime = expiry_datetime.replace(tzinfo=timezone.utc)

    # 2) If they passed a date (no time), lift it to a datetime
    elif isinstance(expiry_datetime, date) and not isinstance(expiry_datetime, datetime):
        expiry_datetime = datetime.combine(
            expiry_datetime,
            time(0, 0, tzinfo=timezone.utc)
        )

    # Now expiry_datetime is guaranteed to be a datetime:
    # 3) Force it to 07:00:00 UTC
    expiry_datetime = expiry_datetime.astimezone(timezone.utc).replace(
        hour=7,
        minute=0,
        second=0,
        microsecond=0,
        tzinfo=timezone.utc
    )

    # 4) Format as ISO with milliseconds and trailing "Z"
    #    e.g.: "2022-01-11T07:00:00.000Z"
    iso = expiry_datetime.strftime("%Y-%m-%dT%H:%M:%S.%f")[:-3] + "Z"
    return iso
```

```python
def download_and_save_data(
    api,
    stock,
    exch,
    strike_price,
    right,
    expiry_date,
    start_datetime,
    end_datetime,
    interval,
    export_path
):
    try:
        response = api.get_historical_data(
            interval=interval,
            from_date=start_datetime,
            to_date=end_datetime,
            stock_code=stock,
            exchange_code=exch,
            product_type="options",
            expiry_date=expiry_date,
            right=right,
            strike_price=str(strike_price)
        )
        if response and "Success" in response:
            df = pd.DataFrame(response["Success"])
            # e.g. df.to_csv(f"{export_path}/{stock}_{expiry_date}_{strike_price}_{right}.csv")
            return True, df
        else:
            return False, None
    except Exception as e:
        print(f"Error fetching data for {strike_price} {right}: {e}")
        return False, None
```

```python
def fetch_data(
    api,
    stock,
    exch,
    expiry_date,
    start_datetime,
    end_datetime,
    start_strike,
    end_strike,
    step,
    max_threads=5,
    interval="1day",
    export_path="HistoricData"
):
    """
    Fetches historical options data for the given strikes and expiry,
    returns a concatenated DataFrame of all successful downloads.
    """
    # 1) Normalize expiry date
    expiry_iso = format_expiry_date(expiry_date)
    strikes = range(start_strike, end_strike + 1, step)

    def run_download_and_collect(strikes_to_run):
        collected = []
        with ThreadPoolExecutor(max_workers=max_threads) as executor:
            # submit all call/put tasks
            futures = {
                executor.submit(
                    download_and_save_data,
                    api,
                    stock,
                    exch,
                    strike,
                    right,
                    expiry_iso,
                    start_datetime,
                    end_datetime,
                    interval,
                    export_path
                ): (strike, right)
                for strike in strikes_to_run
                for right in ("call", "put")
            }
            for future, (strike, right) in futures.items():
                success, df = future.result()
                if success and isinstance(df, pd.DataFrame):
                    # annotate for traceability
                    df = df.copy()
                    df['strike'] = strike
                    df['right'] = right
                    collected.append(df)
        return collected

    # 2) First pass
    all_dfs = run_download_and_collect(strikes)

    # 3) Retry failures once
    #    we detect failures by comparing counts: expected vs obtained
    expected = {(s, r) for s in strikes for r in ("call", "put")}
    got = {(df['strike'].iloc[0], df['right'].iloc[0]) for df in all_dfs}
    failed = [s for (s, r) in expected.difference(got) for _ in (r,)]
    if failed:
        retry_dfs = run_download_and_collect(failed)
        all_dfs.extend(retry_dfs)

    # 4) Concatenate and return
    if all_dfs:
        return pd.concat(all_dfs, ignore_index=True)
    else:
        return pd.DataFrame()
```
